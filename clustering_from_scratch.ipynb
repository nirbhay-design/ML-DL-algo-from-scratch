```python
import random
import string
def random_Hex_generator(n):

  return ['#'+''.join(random.sample(list(string.hexdigits),6)) for i in range(n)]
random_Hex_generator(5)
```

```python
def get_dist(a1,a2):
  return np.sqrt(np.sum((a1-a2) ** 2 , axis = 0))
def K_means(k_clusters,data):
  n_clusters = k_clusters
  cluster_cent = [np.array(data.iloc[i,:]) for i in range(n_clusters)] 
  clusters = []
  cls_ind = []
  for i in range(15):
    iter_clusters = []
    same_cluster_ind = [[] for i in range(n_clusters)]
    for j in range(data.shape[0]):
      arr = np.array(data.iloc[j,:])
      d = np.array([get_dist(cluster_cent[per],arr) for per in range(n_clusters)])
      min_ind = np.argmin(d,axis=0) # getting minimum cluster index
      same_cluster_ind[min_ind].append(j)
      iter_clusters.append(min_ind+1)
    clusters.append(np.array(iter_clusters))
    if(i==14):
      cls_ind = same_cluster_ind
    print(f"clustering after {i+1} iterations: ",same_cluster_ind)
    # updating centeroids 
    cluster_cent = [np.sum(np.array(data.iloc[same_cluster_ind[nirbhay] , :]),axis=0) / len(same_cluster_ind[nirbhay]) for nirbhay in range(n_clusters)]  
  print("\n")
  print("final clusters of all the data points : ",np.array(clusters[-1]))
  print("\n")
  print('clusters final ceteroids are',np.array(cluster_cent))
  data['clusters'] = clusters[-1]
  colors = random_Hex_generator(n_clusters)
  data['c'] = data['clusters'].map({i+1:colors[i] for i in range(n_clusters)})
  from itertools import combinations

  l = combinations([0,1,2,3],2)

  for i, j in list(l):
    plt.scatter(data.iloc[:,i],data.iloc[:,j],c=data.c)
    plt.xlabel(f'feature {i}')
    plt.ylabel(f'feature {j}')
    plt.show()

  lis = combinations([0,1,2,3],3)

  for i,j,k in lis:
    fig = plt.figure(figsize=(26,6))
    ax = fig.add_subplot(131, projection='3d')
    ax.scatter(data.iloc[:,i], data.iloc[:,j], data.iloc[:,k], c=data.c)
    ax.set_xlabel(f'feature {i}')
    ax.set_ylabel(f'feature {j}')
    ax.set_zlabel(f'feature {k}')
    plt.show()
```

```python
K_means(3,data.copy())
```
